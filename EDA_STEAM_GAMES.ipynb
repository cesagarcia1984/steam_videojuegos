```python
import pandas as pd 
# se importa el archivo creado en la subcarpeta ETL y se crea un dataframe 
steam_games = pd.read_csv("D:/Users/Cesar/Desktop/Proyecto_individial_1/Data_ingeneria/ETL/steam_games_ETL")
steam_games.replace({None: pd.NA}, inplace=True)
steam_games.to_json("steam_games",index=False)
```

```python
# hacemos tambien analisis descriptivo de todas la columnas 

steam_games.describe(include='all')

# podemos obsrvar que para valores numericos como "item_id" y "release_year" tenemos la media,desviacion standar,minimo,
# maximo y ademas los percetiles de cada variable, para las otras variables aparecen NaN ya que son culitativas categorias
# pero por el contrario nos muestra de variable categoricas valores unicos ,valor mas frecuente y la frecuencia del valor
# mas frecuente por ejemplo para la columna "genres"  tiene 883 valores unicos de 28852 filas, el valor mas frecuente es
# [Action] y la freceuncia del valor mas frecuente es 1880,  para variables numericas por ejemplo "item_id" 
# media de alrededor 120,522.4 , desviacion standar aproximada 220,970.3, el minimo es 0 y maximo 2,028,850, lo que nos 
# indica que esta sesgada hacia valores mas bajos esto lo veremos mas abajo con histogramas y grafico de barras 

```

```python
import matplotlib.pyplot as plt

# esta parte del codigo se crea un histograma para mirar la distribucion de los datso y encontarar anomalias en 'item_id'
plt.figure(figsize=(10, 6))
plt.hist(steam_games['item_id'], bins=30, color='skyblue', edgecolor='black') 
plt.title('Histograma de item_id')
plt.xlabel('item_id')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# lo que podemos notar es es que los datos de la columna estan sesgados hacia 0 debido a los cambios relizados al pasar 
# a entero tomo los valores null o NaN a 0 lo cual no afecta las cunsultas de las funciones, se podria dectectar valores
# atipicos para filtrarlos pero debido a que los 0 no son cosiderados atipicos no hace ningun afecto sobre 0 
```
